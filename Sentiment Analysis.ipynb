```python
import pandas as pd
dishes = ['noodle','pizza','roll','burger','taco','pasta','steak','pie','greek','hummus','crab','cake','shrimp','salad']
RegData = pd.DataFrame(columns=(['Restaurant']+dishes))
```

```python
import glob
from nltk import sent_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
import nltk
import numpy as np
import datetime
import string
sid = SentimentIntensityAnalyzer()
rowDict = {}
files = glob.glob("reviews_folder/*.csv")
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stopwords.words('english')
stopwords_list = stopwords.words('english') + list(string.punctuation)

for file in files:
    try:
        print(file)
        test_reviews = pd.read_csv(file,delimiter = '|',header=0,names = ['Overall_rating','Review_date','Review_title','Reviewer_ID','Reviews','restaurant_name','restaurant_rating','restaurant_rating_count','restaurant_food_rating','restaurant_service_rating','restaurant_value_rating'])
        for dish in dishes:
            test_reviews[dish] = np.nan
            test_reviews["DateNew"] = np.nan
            for i in range(0,len(test_reviews)):
                review = test_reviews.loc[i,'Reviews'].lower().split()
                review = [word for word in review if word not in stopwords_list]
                review = ' '.join(review)
                sentences =sent_tokenize(review)
                rev_pos = 0
                count = 0
                for line in sentences:
                    if(line.find(dish)!=-1):
                        rev_pos = rev_pos + sid.polarity_scores(line)['pos'] - sid.polarity_scores(line)['neg']
                        count = count + 1
                if(count != 0):
                    test_reviews.loc[i,dish] = rev_pos/count
            subData = test_reviews[test_reviews[dish].notnull()]
            rowDict[dish] = subData[dish].mean()
            rowDict['Overall_rating'] = test_reviews['Overall_rating'].mean()
            rowDict['restaurant_rating'] = test_reviews['restaurant_rating'].mean()
            rowDict['restaurant_food_rating'] = test_reviews['restaurant_food_rating'].mean()
            rowDict['restaurant_value_rating'] = test_reviews['restaurant_value_rating'].mean()
            rowDict['restaurant_service_rating'] = test_reviews['restaurant_service_rating'].mean()
            rowDict['restaurant_name'] = test_reviews['restaurant_name']
        RegData = RegData.append(rowDict, ignore_index=True)
    except OSError:
        print("Corrupt File = "+file)
RegData.to_csv("output.csv", sep='\t', encoding='utf-8')
```

```python
import nltk
from nltk.corpus import wordnet as wn
food = wn.synset('food.n.02')
foodNames = list(set([w for s in food.closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))

foodMatch = []
for file in files:
    print(file)
    try:
        test_reviews = pd.read_csv(file,delimiter = '|',header=0,names = ['Overall_rating','Review_date','Review_title','Reviewer_ID','Reviews','restaurant_name','restaurant_rating','restaurant_rating_count','restaurant_food_rating','restaurant_service_rating','restaurant_value_rating'])
        for i in range(0,len(test_reviews)):
            for x in foodNames:
                if(test_reviews.loc[i,'Reviews'].lower().find(x)!=-1):
                    foodMatch.append(x)
        freq = nltk.FreqDist(foodMatch)
    except OSError:
        print("Corrupt File = "+file)
```
